```python
import pandas as pd

# Load both CSV files directly from local paths (no URLs)
df_speech = pd.read_csv("fomc_statements_2023_2025.csv")
df_market = pd.read_csv("market_data.csv")

# Merge using the 'date' column
df_merged = pd.merge(df_speech, df_market, on="date", how="inner")

# Preview the merged DataFrame
df_merged.head()

```

```python
df_merged.to_csv("merged_fomc_market_data.csv", index=False)
```
